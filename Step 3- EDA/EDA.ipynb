```python
#columns

col_list = list(df.columns)         #df.columns.to_list()
col_list
```

```python
mis_p=df.isna().sum()/len(df)*100
dict(mis_p.to_dict())
```

```python
#datatype

#float64(3), int64(4), object(13),datetime64[ns]

col_object=[]
col_float=[]
col_integer=[]
col_other=[]



for i in col_list:
    
    if df[i].dtype==object:
        col_object.append(i) 
    elif df[i].dtype==float:
        col_float.append(i)
    elif df[i].dtype in (np.int8,np.int32,np.int64):
        col_integer.append(i)    
    else: 
        col_other.append(i)  
print('object=',col_object)
print()
print('int=',col_integer)  
print()
print('float=',col_float)
print()
print('other datatype column=',col_other)        
```

```python
#categorical and numerical columns

#-----usefull for eda part
#-----usefull for feature enginneering part

categorical_col=list(col_object)
categorical_col
```

```python
numerical_col=col_other+col_float
numerical_col
```
